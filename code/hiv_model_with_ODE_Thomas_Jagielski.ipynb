```python
init = State(r=200, l=0, e=0, v=100) 
```

```python
system = System(gamma = 1.36,
                tau = 0.2,
                mu = 1.36e-3,
                beta = 0.00027,
                alpha = 3.6e-2,
                delta = 0.33,
                pi = 100,
                sigma = 2,
                rho = 0.1,
                t0 = 0,
                t_end = 60,
                dt = 0.1,
                init = state)
```

```python
def update_func(state, t, system):
    R, L, E, V = state
    
    drdt = ((gamma * tau) - (mu * R) - (beta * R * V))
    dldt = ((rho * beta * R * V) - (mu * L)- (alpha * L))
    dedt = (((1-rho) * beta * R * V) + (alpha * L) - (delta * E))
    dvdt = ((pi * E) - (sigma * V))
 
    R += drdt * dt
    L += dldt * dt
    E += dedt * dt
    V += dvdt

    return State(r=R, l=L, e=E, v=V)
```

```python
def run_simulation(system, update_func):
    """Runs a simulation of the system.
        
    system: System object
    update_func: function that updates state
    
    returns: TimeFrame
    """
    unpack(system)
    
    frame = TimeFrame(columns=state.index)
    frame.row[t0] = state
    
    for t in linrange(t0, t_end, dt):
        frame.row[t + dt] = update_func(frame.row[t], t, system)
    
    return frame
```

```python
update_func(state, 1, system)
```

```python
results = run_simulation(system, update_func)
```

```python
plot(results.r)
plot(results.l)
plot(results.e)
```

```python
def slope_func(state, t, system):
    
    R, L, E, V = state
    
    drdt = ((gamma * tau) - (mu * R) - (beta * R * V))
    dldt = ((rho * beta * R * V) - (mu * L)- (alpha * L))
    dedt = (((1-rho) * beta * R * V) + (alpha * L) - (delta * E))
    dvdt = ((pi * E) - (sigma * V))

    return drdt, dldt, dedt, dvdt
```

```python
data, details = run_ode_solver(system, slope_func)
data
```

```python
def plot_results(r, l, v, e):
    """Plot the results of a SIR model.
    
    S: TimeSeries
    I: TimeSeries
    R: TimeSeries
    """
    plot(r, '--', label='Susceptible')
    plot(l, '-', label='Infected')
    plot(v, ':', label='Recovered')
    plot(e, ':', label='Recovered')
    decorate(xlabel='Time (days)',
             ylabel='Fraction of population')
```

```python
plot_results(data.r, data.l, data.v, data.e)
```
